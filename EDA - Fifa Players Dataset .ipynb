```python
#Which country has the most number of players 
import pandas as pd

df = pd.read_csv('fifa_data.csv')

player_count = df['Nationality'].value_counts()

top_country = player_count.idxmax()
top_count = player_count.max()

print(f"The Country With The Most Players is {top_country} With {top_count} Players.")
```

```python
#Plot a bar chart of 5 top countries with the most number of players.
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('fifa_data.csv')

player_count = df['Nationality'].value_counts().head(5)

plt.figure(figsize=(10,6))
player_count.plot(kind='bar', color='skyblue')
plt.title('Top 5 Countries with the Most Players in FIFA Dataset', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Number of Players', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Which player has the highest salary
import pandas as pd

df = pd.read_csv('fifa_data.csv')

highest_salary_player = df.loc[df['Wage'].idxmax()]

print(f"Player with the highest salary: {highest_salary_player['Name']}")
print(f"Salary (Wage): {highest_salary_player['Wage']}")
```

```python
#Plot a histogram to get the salary range of the players.
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('fifa_data.csv')

df['Wage'] = df['Wage'].replace({'€': '', 'K': 'e3', 'M': 'e6'}, regex=True)

df['Wage'] = pd.to_numeric(df['Wage'], errors='coerce')

plt.figure(figsize=(10,6))
plt.hist(df['Wage'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Player Salaries (Wage)', fontsize=16)
plt.xlabel('Salary (Wage)', fontsize=12)
plt.ylabel('Number of Players', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
#Who is the tallest player in the fifa
import pandas as pd

df = pd.read_csv('fifa_data.csv')

df['Height'] = pd.to_numeric(df['Height'], errors='coerce')

df = df.dropna(subset=['Height'])

if df.empty:
    print("No valid height data available.")
else:
    tallest_player = df.loc[df['Height'].idxmax()]

    print(f"The tallest player is {tallest_player['Name']}")
    print(f"Height: {tallest_player['Height']} cm")
```

```python
#Which club has the most number of players
import pandas as pd

df = pd.read_csv('fifa_data.csv')

club_count = df['Club'].value_counts()

top_club = club_count.idxmax()
top_club_count = club_count.max()

print(f"The club with the most players is {top_club} with {top_club_count} players.")
```

```python
#Which foot is most preferred by the players?Draw a bar chart for preferred foot 
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('fifa_data.csv')

foot_preference = df['Preferred Foot'].value_counts()

plt.figure(figsize=(8,6))
foot_preference.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Preferred Foot of Players', fontsize=16)
plt.xlabel('Foot', fontsize=12)
plt.ylabel('Number of Players', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
